# GDP & LIFE EXPECTANCY ANALYSIS

In this project, data on GDP and life expectancy from the World Health Organization is analyzed in order to identify the possible relationship between the GDP and life expectancy of six countries. 

The data includes:
* `Country`: the country studied
* `Year`: year in which the data was recorded
* `Life expectancy at birth (years)`
* `GDP`
  

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
data = pd.read_csv('all_data.csv')
data.head()
```

```python
# Change names of colunns for easier analysis
data.rename(columns={'Country':'country','Year':'year','Life expectancy at birth (years)':'life_expectancy','GDP':'GDP'},inplace=True)
data.head()
```

```python
# Observe summary statistcis 
data.describe(include='all')
```

```python
# Data types and potential null values
data.info()
```

```python
# What countries are included in the study?
data.country.value_counts()
```

### Dataset examination

The dtypes of the variables included in the dataset are as expected and there seem to be no missing data or nul values. The six countries included in the dataset are: Chile, China, Germany, Mexico, United States of America, and Zimbabwe. There are 16 observations for each country, containing the values of life expectancy and GDP every year from 2000 to 2015. 

From the summary statistics, there seems to be a high contrasts between the maximum and minimum values for life expectancy and GDP. In fact, the minimum values for both variables is considerably far from the central distribution. Further exploration of the data should offer more information about this discrepancy. 

```python
# Analysis of Life expectancy by year and country
sns.lineplot(data=data, x='year', y='life_expectancy', hue='country', markers=True, palette='colorblind')
plt.ylabel('Live Expectancy at birth (years)')
plt.xlabel('Year')
plt.title('Evolution of life expectancy across time for 6 different countries')
plt.xlim(left=2000,right=2015)
plt.grid(visible=True, which='major', axis='both', alpha=0.5, linestyle=':')
```

```python
# Analysis of GDP by year and country
sns.lineplot(data=data, x='year', y='GDP', hue='country', markers=True, palette='colorblind')
plt.ylabel('Gross Domestic Product (GDP)')
plt.xlabel('Year')
plt.title('Evolution of GDP across time for 6 different countries')
plt.xlim(left=2000,right=2015)
plt.grid(visible=True, which='major', axis='both', alpha=0.5, linestyle=':')
```

```python
# GDP graph by country excluding USA, China, and Germany to be able to evaluate differences in lower income countries
data_lowGDP = data[(data.country != 'China') & (data.country != 'Germany') & (data.country != 'United States of America') & (data.country != 'Mexico')]

ax = plt.subplot()
sns.lineplot(data=data_lowGDP, x='year', y='GDP', hue='country', markers=True, palette='colorblind')
plt.ylabel('Gross Domestic Product (GDP)')
plt.xlabel('Year')
plt.title('Evolution of GDP across time for developing countries')
plt.xlim(left=2000,right=2015)
ax.set_ylim([2.5e+07,3.0e+11])
plt.grid(visible=True, which='major', axis='both', alpha=0.5, linestyle=':')
```

```python
data_lowGDP2 = data[(data.country != 'China') & (data.country != 'Germany') & (data.country != 'United States of America')]

ax = plt.subplot()
sns.lineplot(data=data_lowGDP2, x='year', y='GDP', hue='country', markers=True, palette='colorblind')
plt.ylabel('Gross Domestic Product (GDP)')
plt.xlabel('Year')
plt.title('Evolution of GDP across time for developing countries')
plt.xlim(left=2000,right=2015)
#ax.set_ylim([2.5e+07,3.0e+11])
plt.grid(visible=True, which='major', axis='both', alpha=0.5, linestyle=':')
```

```python
# GDP graph by country excluding USA, China, and Germany to be able to evaluate differences in lower income countries
data_highGDP = data[(data.country == 'China') | (data.country == 'Germany') | (data.country == 'United States of America') | (data.country == 'Mexico')]

ax = plt.subplot()
sns.lineplot(data=data_highGDP, x='year', y='GDP', hue='country', markers='o', palette='colorblind')
plt.ylabel('Gross Domestic Product (GDP)')
plt.xlabel('Year')
plt.title('Evolution of GDP across time for developed countries')
plt.xlim(left=2000,right=2015)
plt.grid(visible=True, which='major', axis='both', alpha=0.5, linestyle=':')
```
